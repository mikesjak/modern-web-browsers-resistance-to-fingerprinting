```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

def print_confusion_matrix(conf_matrix):
    # Labels
    labels = ["Actual Positive", "Actual Negative"]
    columns = ["Predicted Positive", "Predicted Negative"]

    # Create a DataFrame for better labeling
    df_cm = pd.DataFrame(conf_matrix, index=labels, columns=columns)

    # Plot the confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(df_cm, annot=True, fmt=".2f", cmap="Blues", cbar=False)
    # plt.title("Confusion Matrix (Percentages)")
    plt.ylabel("Actual Class")
    plt.xlabel("Predicted Class")
    plt.tight_layout()
    plt.show()


```

# Confusion Matrix for Complpex Algorithm

```python
# Define the confusion matrix values (percentages)
conf_matrix = np.array([[33.15, 12.82],
                        [54.04, 0.00]])

print_confusion_matrix(conf_matrix)
```

# Confusion Matrix for Naive Algorithm

```python
# Define the confusion matrix values (percentages)
conf_matrix = np.array([[61.07, 0],
                        [38.93, 0]])

print_confusion_matrix(conf_matrix)
```
